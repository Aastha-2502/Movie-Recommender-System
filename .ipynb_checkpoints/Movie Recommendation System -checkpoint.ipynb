```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from datetime import date
import pickle

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')

print(f'Movies - {movies.shape}')
print(f'Credits - {credits.shape}')

data = movies.merge(credits, left_on = ['id','title'], right_on = ['movie_id','title']).drop(
    columns = ['movie_id'])

data.shape
```

```python
# For content based recommnedation system -  genres, id, keywords, original_language, overview,
# popularity, production_companies, release_date, title, vote_average, cast, crew
data = data[['id','title','genres','keywords','original_language','overview','popularity','production_companies',
     'release_date','vote_average','cast','crew']]
```

```python
# Null Values
# data.isnull().sum()
# data[data['release_date'].isnull()]
data.dropna(inplace = True)
```

```python
# treating text columns
data['genres'] = data['genres'].apply(lambda x : pd.read_json(x)['name'].to_list() if len(x)>2 else x)
```

```python
data['keywords'] = data['keywords'].apply(lambda x : pd.read_json(x)['name'].to_list() if len(x)>2 else x)
```

```python
data['production_companies'] = data['production_companies'].apply(lambda x : pd.read_json(x)['name'].to_list() if len(x)>2 else x)
```

```python
data['cast'] = data['cast'].apply(lambda x : pd.read_json(x)[:3]['name'].to_list() if len(x)>2 else x)
```

```python
data['crew'] = data['crew'].apply(lambda x : [i['name'] for i in json.loads(x) if i['job'] == 'Director'] if len(x)>2 else x)
```

```python
data['overview'] = data['overview'].str.split()
```

```python
data['genres'] = data['genres'].apply(lambda x : [i.replace(' ','') for i in x])
data['keywords'] = data['keywords'].apply(lambda x : [i.replace(' ','') for i in x])
data['overview'] = data['overview'].apply(lambda x : [i.replace(' ','') for i in x])
data['production_companies'] = data['production_companies'].apply(lambda x : [i.replace(' ','') for i in x])
data['cast'] = data['cast'].apply(lambda x : [i.replace(' ','') for i in x])
data['crew'] = data['crew'].apply(lambda x : [i.replace(' ','') for i in x])
```

```python
data['tags'] = data['overview'] + data['genres'] + data['keywords'] + data['production_companies'] + data['cast'] + data['crew']
```

```python
data.drop(columns = ['genres','keywords','overview','production_companies','cast','crew'], inplace = True)
```

```python
data['tags'] = data['tags'].apply(lambda x : [i for i in x if i not in ['[',']']])
```

```python
ps = PorterStemmer() # convert all the similar words into its original root word.
data['tags'] = data['tags'].apply(lambda x : ' '.join([ps.stem(i).lower() for i in x]))                   
```

```python
# numeric columns
plt.figure()
sns.displot(data['popularity'])

plt.figure()
sns.distplot(data['vote_average'])

plt.figure()
plt.pie(data['original_language'].value_counts(),
        labels = data['original_language'].value_counts().index, shadow = True, 
        labeldistance = 0.4, explode = np.random.rand(data['original_language'].nunique()).tolist())
plt.show()

# as visible, popluarity is highly skewed can be included, even similar votes can also be included
# since original_language is highly imbalance, let's drop this.
```

```python
#data.drop(columns = 'original_language', inplace = True)
```

```python
data['movie_age'] = date.today().year - pd.to_datetime(data['release_date']).dt.year
```

```python
data.drop(columns = 'release_date', inplace = True)
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english') # bags_of_words approach of vectorising
tfid = TfidfVectorizer(max_features = 5000, stop_words = 'english')
vectors = cv.fit_transform(data['tags']).toarray()
vec = tfid.fit_transform(data['tags']).toarray()

num_cols = data[['popularity','vote_average','movie_age']].values
scaler = StandardScaler()
normalise_num_cols = scaler.fit_transform(num_cols)

comb_data = np.hstack((vectors,normalise_num_cols))
combined_data = np.hstack((vec,normalise_num_cols))
```

```python
cosine_sim_matrix = cosine_similarity(comb_data)
```

```python
cosine_matrix_tfid = cosine_similarity(combined_data)
```

```python
cosine_matrix_text_tfid = cosine_similarity(vec)
```

```python
cosine_matrix_text_bow = cosine_similarity(vectors) 
```

```python
def get_recommendation(movie,cosine_sim_matrix,n=5):
    movie_index = data[data['title'] == movie].index[0]
    sim_scores = list(enumerate(cosine_sim_matrix[movie_index]))
    sim_scores = sorted(sim_scores, key = lambda x : x[1], reverse = True)
    sim_scores = sim_scores[1:n+1]
    movie_indices = [i[0] for i in sim_scores]
    return data.iloc[movie_indices,1]
```

```python
get_recommendation('Captain America: Civil War',cosine_sim_matrix)
```

```python
get_recommendation('Captain America: Civil War',cosine_matrix_tfid)
```

```python
get_recommendation('Avatar',cosine_matrix_text_tfid)
```

```python
get_recommendation('Avatar',cosine_matrix_text_bow)
```
